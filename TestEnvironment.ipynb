# Test- and Demo Environment for SERPENS

This notebook is both meant as a means of testing functionality, as well as demonstrating various functionalities.

Import necessary packages:

```python
import matplotlib

from serpens_simulation import SerpensSimulation
from src.species import Species
from serpens_analyzer import SerpensAnalyzer
```

## Setting up and Starting a Simulation

In the following, we initialize a SERPENS simulation and using the predefined HD-189733 b I exoplanet-exomoon system. We add another sourcing exomoon using the `add` method.

```python
ssim = SerpensSimulation(system="HD-189733")
ssim.add(m=8.8e+22, a=267868894.98, r=1820000.0, primary="planet", source=True, 
         species=Species('H2', n_th=0, n_sp=80, mass_per_sec=10**4.8, model_smyth_v_b=0.95*1000, model_smyth_v_M=15.24*1000, 
                         lifetime=4*60, beta=0)
         )
```

Run time simulation using `advance`:

```python
ssim.advance(20, verbose=False)
```

## Analyzing the Results

Using the `SerpensAnalyzer` class, we can plot results from the previous simulation.

```python
%matplotlib inline

sa = SerpensAnalyzer(save_output=False, reference_system="source0")
_ = sa.plot_planar(timestep=20, d=3, scatter=True, triplot=False, trialpha=1,
                   lim=8, celest_colors=['orange', 'sandybrown', 'red', 'gainsboro'],
                   colormap=matplotlib.colormaps["afmhot"], show_source=True, show=True, single_plot=False, figsize=10)
```
